```r
library(ggplot2)
library(ggpubr)


options(repr.plot.width = 10, repr.plot.height = 6)
EffectNames= c("250","100","50","25","5")
MotifNames = annotations[annotations$Timepoint=="5",][order(annotations[annotations$Timepoint=="5",]$Total, decreasing=FALSE),"Motif"]
ggballoonplot(annotations,  x = "Timepoint", y = "Motif", fill = "Recall", size="Total")+
  scale_fill_viridis_c(option = "C") +   scale_x_discrete(limits = EffectNames)  +   scale_y_discrete(limits = MotifNames)

ggsave("subfigs/hitcalling_compare/recall_with_read_depth.pdf", device = "pdf",  width = 5, height = 6, dpi = 300)
```
